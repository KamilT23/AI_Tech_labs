Пропусков нет в данных. Добавим пропущенных значений

```python
def nan_val(data, column, nan_prob):
    data_with_nan = data.copy()
    mask = np.random.random(data.shape[0]) <= nan_prob
    data_with_nan.loc[mask, column] = np.nan
    return data_with_nan
```

```python
column = 'height'
df_nan = nan_val(df, column, nan_prob = 0.05)
```

```python
df_nan.head(40)
```

Работа с пропущенными значениями: Удаление строк с пропущенными значениями:

Заполнение пропущенных значений средним значением по колонке (например, для колонки 'height'):

```python
values = {column: df[column].mean()}
df_mean = df_nan.fillna(value = values)
```

```python
print(f'Кол-во строк с пропущ. данными до заполения средними: {df_nan.isna().sum(axis=1).sum()}')
print(f'Кол-во строк с пропущ. данными после заполения средними: {df_mean.isna().sum(axis=1).sum()}')
```

Посчитаем индекс массы тела (bmi) вычисляется как отношение массы тела к квадрату роста в метрах

```python
df['bmi'] = (df['weight']) / ((df['height'] / 100) ** 2)
df.head()
```

Создание нового поля через DataFrame.apply на основе weight и height

```python
df = df.drop('bmi', axis = 1)
```

```python
def calculate_bmi(row):
    weight_kg = row['weight']
    height_cm = row['height']  # Изменяем высоту в сантиметры
    height_m = height_cm / 100  # Переводим в метры
    bmi = weight_kg / (height_m ** 2)
    return bmi

# Применяем функцию к каждой строке DataFrame
df['bmi'] = df.apply(lambda x: calculate_bmi(x), axis=1)
```

```python
df.head(15)
```

С помощью Series.apply

```python
# Функция для определения здорового образа жизни
def is_healthy_lifestyle(row):
    if row['bmi'] < 25 and row['cholesterol'] == 1 and row['gluc'] == 1 and row['smoke'] == 0 and row['alco'] == 0 and row['active'] == 1:
        return True
    else:
        return False

# Добавление нового столбца "healthy_lifestyle" с использованием Series.apply
df['healthy_lifestyle'] = df.apply(lambda x: is_healthy_lifestyle(x), axis=1)
```

**Вычислить квантили распределения какого-либо вещественного признака (с использованием numpy.quantile или numpy.percentile).**

```python
# Вычисление квантилей для вещественного признака (например, 'age' или 'bmi')
feature = 'age'

# Вычисление квантилей (25%, 50% и 75%)
q1 = np.percentile(df[feature], 25)
median = np.percentile(df[feature], 50)
q3 = np.percentile(df[feature], 75)

print(f"25% (Q1): {q1}")
print(f"Median (Q2): {median}")
print(f"75% (Q3): {q3}")
```

**Вычислить (в виде текста) гистограмму какого-либо вещественного признака** (с использованием numpy.histogram). Значения гистограммы можно использовать в качестве загрубленного числового признака для заданий (5) или (6).

```python
# Вычисление гистограммы
hist, bin_edges = np.histogram(df[feature], bins=5)

# Представление гистограммы в текстовом виде
histogram_text = "Histogram for '{}' feature:\n".format(feature)
for i in range(len(hist)):
    histogram_text += "[{:.2f} - {:.2f}]: {}\n".format(bin_edges[i], bin_edges[i + 1], hist[i])

print(histogram_text)
```

**Проитерировать DataFrame построчно .iterrows() и выполнить какую-либо операцию внутри цикла.**

```python
# Создаем переменные для хранения суммы веса и количества женщин
total_weight = 0
num_female = 0

# Проходим построчно по DataFrame
for index, row in df.iterrows():
    if row['gender'] == 1:  # 1 - это женщина
        total_weight += row['weight']
        num_female += 1


average_weight_female = total_weight / num_female
print("Средний вес женщин:", average_weight_female)
```

**Продемонстировать работу .merge.**

```python
data1 = {
    'age': [25, 30, 35],
    'name': ['Alice', 'Bob', 'Charlie']
}
df1 = pd.DataFrame(data1)

data2 = {
    'age': [25, 35, 40],
    'city': ['London', 'Paris', 'Berlin']
}
df2 = pd.DataFrame(data2)

# Объединяем датасеты по столбцу 'age'
merged_df = pd.merge(df1, df2, on='age', how='inner')
```
